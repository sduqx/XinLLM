# 练习 2.1 未知词的字节对编码

尝试使用 tiktoken 库中的 BPE 分词器对未知单词 "Akwirw ier" 进行处理，并输出各个token ID。接着，对此列表中的每个结果整数调用 decode 函数，以重现token id到token的映射。最后，调用token ID 的 decode 方法，检查它是否能够重建原始输入 "Akwirw ier"。


```python
import tiktoken
tokenizer = tiktoken.get_encoding("gpt2")
integers = tokenizer.encode("Akwirw ier")
print(integers)
```

```python
for integer in integers:
    print(f"{integer} -> {tokenizer.decode([integer])}")
```

```python
tokenizer.decode(integers)
```
